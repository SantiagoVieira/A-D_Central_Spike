```python
# Correr primero la celda de arriba
# ejercicio 4 - punto 6 - Codigo sin error al momento de calcular n = 151145018

import gmpy2

nombre = "sara"
longitud_nombre = len(nombre)

k1 = gmpy2.mpfr("1830143396396800") / gmpy2.mpfr("186121700512702523014509")
k2 = gmpy2.mpfr("38127987424931") / gmpy2.mpfr("186121700512702523014509")
r = gmpy2.mpfr("3")
n = gmpy2.mpfr("151145018")

# Calcular el resultado
result = k1 * r**n + k2 * n * (-r)**n
print(result)

#1.3293116428022114e+72114499
```
